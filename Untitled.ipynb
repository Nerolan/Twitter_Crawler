# Stemming

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

def tokenization(text):
    text = re.split('\W+', text)
    return text

stopword = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

ps = nltk.PorterStemmer()
def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

def pos_tagging(listOfTweets):
    listOfTaggedTweets = [nltk.pos_tag(tweet) for tweet in listOfTweets]
    return listOfTaggedTweets
```

```python
df['Tweet_punct'] = df['Tweets'].apply(lambda x: remove_punct(x))
df['Tweet_tokenized'] = df['Tweet_punct'].apply(lambda x: tokenization(x.lower()))
df['Tweet_nonstop'] = df['Tweet_tokenized'].apply(lambda x: remove_stopwords(x))
df['Tweet_tagged'] = df['Tweet_nonstop'].apply(lambda x: pos_tagging(x))
df['Tweet_stemmed'] = df['Tweet_nonstop'].apply(lambda x: stemming(x))
df['Tweet_lemmatized'] = df['Tweet_nonstop'].apply(lambda x: lemmatizer(x))
```

```python
df['Tweet_tagged'] = df['Tweets'].apply(lambda x: pos_tagging(x))
```

# Term Frequency & Inverse Data Frequency

```python
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit_transform(df['Tweet_punct'])
```

```python
df['Tweet_IDF'] = [vectorizer.idf_ for entry in df['Tweet_lemmatized']]
```

```python
df.head()
```

```python
df['Features'] = vectorizer.get_feature_names()
```

```python
def calcTF(wordDict, lemmatizedTweet):
    tfDict = {}
    numberOfWords = len(lemmatizedTweet)
    for word, count in wordDict.items():
        tfDict[word] = count / float(numberOfWords)
    return tfDict
```

```python
for cell in df['Tweet_TF']:
    df.at[cell, 'Tweet_TF'] = calcTF(df.at[cell, 'Tweet_lemmatized'])
```
